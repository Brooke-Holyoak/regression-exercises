```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from pydataset import data

from statsmodels.formula.api import ols

from scipy import stats

from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression 
from sklearn.metrics import mean_squared_error
from math import sqrt


import warnings
warnings.filterwarnings('ignore')
```

```python
#Import tips dataset from pydataset

tips = data('tips')
```

```python
#Make sure to look at the df
tips.head()
```

```python
#Fit a linear regression model (ordinary least squares) and compute yhat, 
#predictions of tip using total_bill
#target is y, feature to predict is x
```

```python
#First step is to define x and y.

X = tips.total_bill
y = tips.tip
```

```python
#Next create a baseline prediction from the mean of y: this is the yhat baseline
#Add a column called yhat_baseline that contains the baseline prediction

tips['yhat_baseline'] = y.mean()
```

```python
#Look at the df to check

tips.head()
```

```python
# defining and creating the model and model predictions

# generate parameters, i.e. create model
ols_model = LinearRegression().fit(tips[['total_bill']], y)

# compute predictions and add to original dataframe
tips['yhat'] = ols_model.predict(tips[['total_bill']])

#look at df
tips.head()
```

```python
# compute residuals

tips['residual'] = tips['yhat'] - y
tips['residual_baseline'] = tips['yhat_baseline'] - y
```
